```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0,5,15)
y= x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r','o')
```

```python
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('Bhavya')
axes.set_ylabel('Sodai')
axes.set_title('Distance')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Smiley')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
#axes.plot(x,y)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
  current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('Why')
axes[1].plot(y,x)
axes[1].set_ylabel('NO sleep')
plt.tight_layout()
```

- figure size and DPI

```python
fig = plt.figure(figsize = (5,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(5,2))
axes[0].plot(x,y)
axes[0].set_ylabel('OMG')
axes[1].plot(y,x)
fig.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X^2')
ax.plot(x,x**3,label='X^3')
ax.legend(loc='best')
#we can also give in tuples
# specifying percentage of fiure from the bottom left corner
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange')#you can also put RGB hex codes
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='--',marker='o',markersize=10,
        markerfacecolor='orange',markeredgewidth=3,markeredgecolor='blue')
ax.set_xlim([1,4])
```
